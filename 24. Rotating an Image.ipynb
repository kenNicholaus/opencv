```python
import cv2
import matplotlib.pyplot as plt

def main():
    
    path = "D:\\kenneth\\opencv\\dataset\\"
    
    imgpath1 =  path + "4.1.04.tiff"
    imgpath2 =  path + "4.1.01.tiff"
    img1 = cv2.imread(imgpath1, 1)
    img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
    
    
    rows, columns, channels = img1.shape
    
    #R = cv2.getRotationMatrix2D((columns/2, rows/2), 400, 1)
    R = cv2.getRotationMatrix2D((columns/2, rows/2), 200, 1)
    
    print(R)
    
    
    output = cv2.warpAffine(img1, R, (columns, rows))
    
    plt.imshow(img1)
    plt.title('Original')
    plt.show()
    
    
    plt.imshow(output)
    plt.title('Rotated Image')
    plt.show()

if __name__ == "__main__":
    main()
```
