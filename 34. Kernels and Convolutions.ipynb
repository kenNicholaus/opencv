```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def main():
    
    path = "C:\\Users\\kenneth\\opencv\\dataset\\"
    
    imgpath1 =  path + "4.1.04.tiff"
    imgpath2 =  path + "4.1.01.tiff"
    img = cv2.imread(imgpath1, 1)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    
#     k = np.array(np.ones((11, 11), np.float32))/121
    
#     k = np.array(([0, -1, 0], [-1, 5, -1], [0, -1, 0]), np.float32)
    
    k = np.array(([1, 0, -1], [0, 0, 0], [-1, 0, 1]), np.float32)
    
    k1 = np.array(([0, 1, 0], [-1, 5, -1], [0, -1, 0]), np.float32)
    k2 = np.array(np.ones((11, 11), np.float32))/121
    
    print('edge detection: ', k)
    print('sharpen: ', k1)
    

    output = cv2.filter2D(img, -1, k)
    output1 = cv2.filter2D(img, -1, k1)
    
    plt.subplot(1, 3, 1)
    plt.imshow(img)
    plt.title('Original Image')
    
    plt.subplot(1, 3, 2)
    plt.imshow(output)
    plt.title('Edge Detection')
    
    plt.subplot(1, 3, 3)
    plt.imshow(output1)
    plt.title('Sharpen')
    
    plt.show()

if __name__ == "__main__":
    main()
```
