```python
import cv2
import matplotlib.pyplot as plt

def main():
    
    path = "C:\\Users\\kenneth\\opencv\\dataset\\"
    
    imgpath1 =  path + "4.1.04.tiff"
    imgpath2 =  path + "4.1.01.tiff"
    
    img1 = cv2.imread(imgpath1, 1)
    img2 = cv2.imread(imgpath2, 1)
    
    img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
    img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
    
    alpha = 0.5
    beta = 0.5
    gamma = 0
    
    output = cv2.addWeighted(img1, alpha, img2, beta, gamma)
    
    titles = ['Liquid Drop', 'Lena', 'Weighted Addition']
    images = [img1, img2, output]
    
    print()
    
    for i in range(3):
        plt.subplot(1, 3, i+1)
        plt.imshow(images[i])
        plt.title(titles[i])
        plt.xticks([])
        plt.yticks([])

    plt.show()  
 
if __name__ == "__main__":
    main()
```

```python
import cv2
import numpy as np
import time
#import matplotlib.pyplot as plt

def main():
    
    path = "C:\\Users\\kenneth\\opencv\\dataset\\"
    
    imgpath1 =  path + "4.1.04.tiff"
    imgpath2 =  path + "4.1.01.tiff"
    
    img1 = cv2.imread(imgpath1, 1)
    img2 = cv2.imread(imgpath2, 1)
    
    img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
#    img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
    
    for i in np.linspace(0, 1, 200):
        alpha = i
        beta = 1 - alpha
        output = cv2.addWeighted(img1, alpha, img2, beta, 0)
        cv2.imshow('Transition', output)
        time.sleep(0.02)
        if cv2.waitKey(1) == 27:
            break
        
    cv2.destroyAllWindows()
 
if __name__ == "__main__":
    main()
```
