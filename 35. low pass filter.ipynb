```python
import cv2
import matplotlib.pyplot as plt


def main():
    
    path = "C:\\Users\\kenneth\\opencv\\dataset\\"
    
    imgpath1 =  path + "4.1.07.tiff"
    imgpath2 =  path + "4.1.01.tiff"
    img = cv2.imread(imgpath1, 1)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    
    box = cv2.boxFilter(img, -1, (53, 53))
    
    blur = cv2.blur(img, (13, 13))
    
    gaussian = cv2.GaussianBlur(img, (37, 37), 0)
    
    titles = ['Original Image', 'Box Filter', 
              'Blur', 'Gaussian Blur']

    outputs = [img, box, blur, gaussian]
    
    
    for i in range(4):
        plt.subplot(2, 2, i+1)
        plt.imshow(outputs[i])
        plt.title(titles[i])
        plt.xticks([])
        plt.yticks([])
    plt.show()

if __name__ == "__main__":
    main()
```
