```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def main():
    
    path = "C:\\Users\\ken\\opencv\\dataset\\"
    
    imgpath1 =  path + "4.1.04.tiff"
    imgpath2 =  path + "4.1.01.tiff"
    img = cv2.imread(imgpath1, 0)

    plt.subplot(1, 2, 1)
    plt.imshow(img, cmap='gray')
    plt.title('Image')
    plt.xticks([])
    plt.yticks([])
    
    plt.subplot(1, 2, 2)
    hist, bins = np.histogram(img.ravel(), 256, [0,255])
    plt.xlim([0, 255])
    plt.ylim([0, 3000])
    plt.plot(hist)
    plt.title('Histogram')
    plt.show()

if __name__ == "__main__":
    main()
```
