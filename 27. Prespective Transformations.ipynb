```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def main():
    
    path = "D:\\kenneth\\opencv\\dataset\\"
    
    imgpath1 =  path + "4.1.04.tiff"
    imgpath2 =  path + "4.1.01.tiff"
    img1 = cv2.imread(imgpath2, 1)
    img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
    
    
    rows, columns, channels = img1.shape
    
    point1 = np.float32([[0, 0], [400, 0], [0, 400], [400, 400]])
    point2 = np.float32([[0, 0], [500, 0], [0, 500], [500, 500]])
    
    P = cv2.getPerspectiveTransform(point1, point2)
    
    print(P)
    
    
    changed = cv2.warpPerspective(img1, P, (250, 250))
    
#     titles = ['Original Image', 'Changed Perspective']
#     outputs = ['img1','changed']
    
 
    plt.imshow(img1)
    plt.title('original')
    plt.show()
    
    
    plt.imshow(changed)
    plt.title('Changed Perspective')
    plt.show()

if __name__ == "__main__":
    main()
```
