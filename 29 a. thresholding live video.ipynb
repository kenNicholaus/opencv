```python
import cv2

def main():
    
    windowName = ['Binary', 'Binary Inv', 'Zero', 'Zero Inv', 'Trunc']
    
    cap = cv2.VideoCapture(0)
    
    if cap.isOpened():
        ret, frame = cap.read()
    else:
        ret = False


    while ret:
    
        ret, frame = cap.read()
        
        th = 127
        max_val = 255
    
        ret, o1 = cv2.threshold(frame, th, max_val, cv2.THRESH_BINARY)
        ret, o2 = cv2.threshold(frame, th, max_val, cv2.THRESH_BINARY_INV)
        ret, o3 = cv2.threshold(frame, th, max_val, cv2.THRESH_TOZERO)
        ret, o4 = cv2.threshold(frame, th, max_val, cv2.THRESH_TOZERO_INV)
        ret, o5 = cv2.threshold(frame, th, max_val, cv2.THRESH_TRUNC)
#         th1 = 0
  
    
#         ret, o6 = cv2.threshold(frame, th1, max_val, cv2.THRESH_BINARY + cv2.THRESH_OTSU )
#         ret, o7 = cv2.threshold(frame, th1, max_val, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU )
#         ret, o8 = cv2.threshold(frame, th1, max_val, cv2.THRESH_TOZERO + cv2.THRESH_OTSU )
#         ret, o9 = cv2.threshold(frame, th1, max_val, cv2.THRESH_TOZERO_INV + cv2.THRESH_OTSU )
#         ret, o10 = cv2.threshold(frame, th1, max_val, cv2.THRESH_TRUNC + cv2.THRESH_OTSU )
        
        cv2.imshow(windowName[0], o1)
        cv2.imshow(windowName[1], o2)
        cv2.imshow(windowName[2], o3)
        cv2.imshow(windowName[3], o4)
        cv2.imshow(windowName[4], o5)
#         cv2.imshow(windowName[5], o6)
#         cv2.imshow(windowName[6], o7)
#         cv2.imshow(windowName[7], o8)
#         cv2.imshow(windowName[8], o9)
#         cv2.imshow(windowName[9], o10)
        
        
        
        
        if cv2.waitKey(1) == 27:
            break

    cv2.destroyAllWindows()
    cap.release()

if __name__ == "__main__":
    main()
```
